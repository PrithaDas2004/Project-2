```python
import os
import random
import json
import librosa
import numpy as np
import pandas as pd
from collections import defaultdict
from transformers import pipeline
import whisper
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

class ESC50Analyzer:
    def __init__(self, train_new_model=False):
        # Load ESC-50 urgency mappings from CSV
        self.urgency_map = self.load_urgency_mappings('/content/drive/MyDrive/sound_urgency_esc50.csv')

        # Initialize ESC-50 class mappings
        self.esc50_classes = [
            'dog', 'rooster', 'pig', 'cow', 'frog', 'cat', 'hen', 'insects', 'sheep', 'crow',
            'rain', 'sea_waves', 'crackling_fire', 'crickets', 'chirping_birds', 'water_drops',
            'wind', 'pouring_water', 'toilet_flush', 'thunderstorm', 'crying_baby', 'sneezing',
            'clapping', 'breathing', 'coughing', 'footsteps', 'laughing', 'brushing_teeth',
            'snoring', 'drinking_sipping', 'door_wood_knock', 'mouse_click', 'keyboard_typing',
            'door_wood_creaks', 'can_opening', 'washing_machine', 'vacuum_cleaner', 'clock_alarm',
            'clock_tick', 'glass_breaking', 'helicopter', 'chainsaw', 'siren', 'car_horn', 'engine',
            'train', 'church_bells', 'airplane', 'fireworks', 'hand_saw'
        ]

        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.esc50_classes)}
        self.idx_to_class = {idx: cls for idx, cls in enumerate(self.esc50_classes)}

        # Call initialize_models to setup models
        self.initialize_models(train_new_model)

    def load_urgency_mappings(self, csv_path):
        """Load urgency mappings from CSV with default fallback"""
        try:
            # Try to load from CSV (assuming format: Sound,Urgency)
            df = pd.read_csv(csv_path)
            return dict(zip(df['Sound'], df['Urgency']))
        except Exception as e:
            print(f"Error loading urgency mappings: {e}")
            # Default mappings for ESC-50 critical classes
            default_map = {
                'glass_breaking': 'high',
                'siren': 'high',
                'car_horn': 'high',
                'fireworks': 'high',
                'chainsaw': 'high',
                'helicopter': 'medium',
                'crying_baby': 'high',
                'thunderstorm': 'high',
                '_default': 'low'
            }
            print("Using default urgency mappings")
            return default_map

    def get_urgency(self, class_name):
        """Get urgency level for a class with fallback to default"""
        return self.urgency_map.get(class_name, self.urgency_map.get('_default', 'low'))

    def initialize_models(self, train_new_model):
        """Initialize sound classification and emotion analysis models"""
        # Initialize emotion classifier
        self.emotion_classifier = pipeline(
            "text-classification",
            model="j-hartmann/emotion-english-distilroberta-base",
            top_k=1
        )

        # Initialize speech-to-text model
        self.stt_model = whisper.load_model("base")

        self.scaler = StandardScaler()

        if train_new_model:
            print("Training new ESC-50 classification model...")
            self.train_sound_classifier()
        else:
            self.load_sound_model()

    def load_sound_model(self):
        """Try to load ESC-50 classification model"""
        model_paths = [
            "esc50_classifier.h5",
            "./esc50_classifier.h5",
            "/content/esc50_classifier.h5",
            "/content/drive/MyDrive/esc50_classifier.h5"
        ]

        for path in model_paths:
            try:
                self.sound_model = load_model(path)
                print(f"ESC-50 classification model loaded from {path}")
                # Load the corresponding scaler
                self.scaler = StandardScaler()
                self.scaler.mean_ = np.load("esc50_scaler_mean.npy")
                self.scaler.scale_ = np.load("esc50_scaler_scale.npy")
                print("Feature scaler loaded successfully")
                return
            except Exception as e:
                print(f"Error loading model from {path}: {str(e)}")
                continue

        print("ESC-50 classification model not found. You may want to train a new one.")

    def audio_augmentation(self, audio, sr):
        """Apply random audio augmentations"""
        # Time stretching
        rate = random.uniform(0.8, 1.2)
        audio = librosa.effects.time_stretch(audio, rate=rate)

        # Pitch shifting
        steps = random.randint(-2, 2)
        audio = librosa.effects.pitch_shift(audio, sr=sr, n_steps=steps)

        # Add noise
        noise = np.random.normal(0, 0.002, audio.shape)
        audio = audio + noise

        return audio * random.uniform(0.5, 1.5)

    def extract_features(self, audio, sample_rate):
        """Extract features optimized for ESC-50 dataset"""
        try:
            if sample_rate != 22050:
                audio = librosa.resample(audio, orig_sr=sample_rate, target_sr=22050)
                sample_rate = 22050

            features = []

            # MFCCs (50 coefficients for ESC-50)
            mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=50)
            mfccs_mean = np.mean(mfccs.T, axis=0)
            mfccs_var = np.var(mfccs.T, axis=0)
            features.extend(mfccs_mean)
            features.extend(mfccs_var)

            # Mel Spectrogram (128 bands)
            mel = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128)
            mel_mean = np.mean(mel.T, axis=0)
            mel_var = np.var(mel.T, axis=0)
            features.extend(mel_mean)
            features.extend(mel_var)

            # Chroma features
            chroma = librosa.feature.chroma_stft(y=audio, sr=sample_rate)
            chroma_mean = np.mean(chroma.T, axis=0)
            chroma_var = np.var(chroma.T, axis=0)
            features.extend(chroma_mean)
            features.extend(chroma_var)

            # Spectral features
            spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sample_rate)[0]
            spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sample_rate)[0]
            spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sample_rate)[0]
            spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sample_rate)

            features.extend([
                np.mean(spectral_centroid), np.var(spectral_centroid),
                np.mean(spectral_bandwidth), np.var(spectral_bandwidth),
                np.mean(spectral_rolloff), np.var(spectral_rolloff),
                np.mean(spectral_contrast), np.var(spectral_contrast)
            ])

            # Temporal features
            rms = librosa.feature.rms(y=audio)[0]
            zcr = librosa.feature.zero_crossing_rate(audio)[0]
            features.extend([np.mean(rms), np.var(rms), np.mean(zcr), np.var(zcr)])

            # Harmonic and percussive components
            harmonic = librosa.effects.harmonic(audio)
            percussive = librosa.effects.percussive(audio)
            features.extend([np.mean(harmonic), np.var(harmonic), np.mean(percussive), np.var(percussive)])

            return np.array(features)

        except Exception as e:
            print(f"Error extracting features: {e}")
            return None

    def load_esc50_data(self, dataset_path, max_samples_per_class=500):
        """Load ESC-50 dataset with augmentation"""
        features = []
        labels = []
        class_counts = defaultdict(int)

        print(f"Loading ESC-50 data from {dataset_path}")

        for class_name in self.esc50_classes:
            class_path = os.path.join(dataset_path, class_name)
            if not os.path.exists(class_path):
                print(f"Warning: Class folder '{class_name}' not found in dataset path")
                continue

            print(f"Processing {class_name}...")
            class_id = self.class_to_idx[class_name]

            audio_files = [f for f in os.listdir(class_path) if f.endswith(('.wav', '.mp3', '.ogg', '.flac'))]
            random.shuffle(audio_files)

            samples_loaded = 0
            for audio_file in audio_files:
                if samples_loaded >= max_samples_per_class:
                    break

                file_path = os.path.join(class_path, audio_file)

                try:
                    audio, sr = librosa.load(file_path, sr=None)

                    # ESC-50 files are 5 seconds long, pad/trim to match
                    target_length = 5 * sr
                    if len(audio) < target_length:
                        audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
                    else:
                        audio = audio[:target_length]

                    # Apply augmentation
                    augmented_audio = self.audio_augmentation(audio, sr)

                    # Extract features for both original and augmented audio
                    for a in [audio, augmented_audio]:
                        feature = self.extract_features(a, sr)
                        if feature is not None:
                            features.append(feature)
                            labels.append(class_id)
                            samples_loaded += 1

                            if samples_loaded >= max_samples_per_class:
                                break

                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")

        return np.array(features), np.array(labels)

    def plot_training_metrics(self, history):
        """Plot training metrics"""
        plt.figure(figsize=(12, 4))

        # Plot training & validation accuracy
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'])
        plt.plot(history.history['val_accuracy'])
        plt.title('Model Accuracy')
        plt.ylabel('Accuracy')
        plt.xlabel('Epoch')
        plt.legend(['Train', 'Validation'], loc='lower right')

        # Plot training & validation loss
        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'])
        plt.plot(history.history['val_loss'])
        plt.title('Model Loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend(['Train', 'Validation'], loc='upper right')

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, y_true, y_pred):
        """Plot confusion matrix for ESC-50 classes"""
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(20, 20))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.esc50_classes,
                    yticklabels=self.esc50_classes)
        plt.title('ESC-50 Confusion Matrix')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.show()

    def train_sound_classifier(self):
        """Train new classifier on ESC-50 dataset"""
        dataset_path = input("Enter path to ESC-50 dataset folder (with class subfolders): ").strip()

        if not os.path.exists(dataset_path):
            print("Dataset folder not found.")
            return

        print("Loading ESC-50 dataset...")
        X, y = self.load_esc50_data(dataset_path)

        if len(X) == 0:
            print("No data loaded. Check dataset path and structure.")
            print("Expected folder structure: dataset_root/class_name/audio_files.wav")
            return

        # Convert labels to categorical
        y_categorical = to_categorical(y, num_classes=len(self.esc50_classes))

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_categorical, test_size=0.2, random_state=42, stratify=y
        )

        # Scale features
        self.scaler = StandardScaler()
        X_train = self.scaler.fit_transform(X_train)
        X_test = self.scaler.transform(X_test)

        # Save scaler parameters
        np.save("esc50_scaler_mean.npy", self.scaler.mean_)
        np.save("esc50_scaler_scale.npy", self.scaler.scale_)

        # Build model - deeper architecture for 50 classes
        model = Sequential([
            Dense(2048, activation='relu', input_shape=(X_train.shape[1],)),
            Dropout(0.5),
            Dense(1024, activation='relu'),
            Dropout(0.5),
            Dense(512, activation='relu'),
            Dropout(0.5),
            Dense(256, activation='relu'),
            Dropout(0.3),
            Dense(128, activation='relu'),
            Dropout(0.3),
            Dense(len(self.esc50_classes), activation='softmax')
        ])

        model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

        print("Training model...")
        history = model.fit(
            X_train, y_train,
            validation_data=(X_test, y_test),
            epochs=200,
            batch_size=128,
            callbacks=[early_stop],
            verbose=1
        )

        # Plot training metrics
        self.plot_training_metrics(history)

        # Generate predictions for evaluation
        y_pred = model.predict(X_test)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)

        # Plot confusion matrix
        self.plot_confusion_matrix(y_true_classes, y_pred_classes)

        # Print classification report
        print("\nClassification Report:")
        print(classification_report(y_true_classes, y_pred_classes, target_names=self.esc50_classes))

        model.save("esc50_classifier.h5")
        self.sound_model = model
        print("ESC-50 model training completed.")

    def detect_sound_events(self, audio_path):
        """Detect sound events using ESC-50 model"""
        events = []
        try:
            audio, sr = librosa.load(audio_path, sr=None)

            # Ensure audio is 5 seconds long (ESC-50 standard)
            target_length = 5 * sr
            if len(audio) < target_length:
                audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
            else:
                audio = audio[:target_length]

            features = self.extract_features(audio, sr)
            if features is not None:
                features = self.scaler.transform(features.reshape(1, -1))
                prediction = self.sound_model.predict(features)
                predicted_class_idx = np.argmax(prediction, axis=1)[0]
                confidence = np.max(prediction)
                detected_class = self.idx_to_class.get(predicted_class_idx)

                if detected_class:
                    events.append({
                        'class': detected_class,
                        'confidence': float(confidence),
                        'urgency': self.get_urgency(detected_class)
                    })
        except Exception as e:
            print(f"Error in sound event detection: {e}")
        return events

    def analyze_emotion(self, text):
        """Analyze emotion from text"""
        if not text.strip():
            return "neutral", 0.0

        try:
            result = self.emotion_classifier(text)
            if result and len(result) > 0:
                emotion = result[0][0]['label']
                confidence = result[0][0]['score']
                return emotion, float(confidence)
        except Exception as e:
            print(f"Error in emotion analysis: {e}")

        return "neutral", 0.0

    def format_output(self, analysis_result):
        """Format the analysis results for display"""
        output = "\n=== ESC-50 Audio Analysis Results ===\n"

        # Sound events
        output += "\n🔊 Detected Sound Events:\n"
        if analysis_result['sound_events']:
            for event in analysis_result['sound_events']:
                output += (f"- {event['class'].replace('_', ' ').title()} "
                          f"[Confidence: {event['confidence']:.2f}, "
                          f"Urgency: {event['urgency'].capitalize()}]\n")
        else:
            output += "No sound events detected\n"

        # Transcription
        if analysis_result['transcription']:
            output += f"\n🗣️ Transcription: \"{analysis_result['transcription']}\""
            output += f"\n😃 Emotion: {analysis_result['emotion']['label'].capitalize()} "
            output += f"(Confidence: {analysis_result['emotion']['confidence']:.2f})\n"
        else:
            output += "\nNo speech detected\n"

        # Insights
        output += f"\n💡 Actionable Insights:\n"
        output += f"- Context: {analysis_result['context']}\n"
        output += f"- Priority: {analysis_result['priority']}\n"

        return output

    def generate_insights(self, sound_events, emotion_result):
        """Generate actionable insights from analysis"""
        insights = []
        context_parts = []
        priority = "Low"

        # Sound-based insights
        if sound_events:
            high_urgency = [e for e in sound_events if e['urgency'] == 'high']
            medium_urgency = [e for e in sound_events if e['urgency'] == 'medium']

            if high_urgency:
                sounds = ", ".join(set(e['class'].replace('_', ' ').title() for e in high_urgency))
                context_parts.append(f"High urgency sounds detected: {sounds}")
                priority = "High"

            if medium_urgency:
                sounds = ", ".join(set(e['class'].replace('_', ' ').title() for e in medium_urgency))
                context_parts.append(f"Medium urgency sounds detected: {sounds}")
                if priority != "High":
                    priority = "Medium"

        # Emotion-based insights
        emotion = emotion_result['label']
        confidence = emotion_result['confidence']

        if confidence > 0.7:
            if emotion in ['anger', 'fear']:
                context_parts.append(f"Strong negative emotion detected ({emotion})")
                if priority != "High":
                    priority = "Medium"
            elif emotion == 'joy':
                context_parts.append("Positive emotional state detected")

        if not context_parts:
            context_parts.append("No significant events detected")

        context = ". ".join(context_parts) + "."

        # Enhanced priority determination
        if priority == "High":
            priority = "Immediate attention required"
        elif priority == "Medium":
            priority = "Monitoring recommended"
        else:
            priority = "No immediate action needed"

        return {
            'context': context,
            'priority': priority
        }

    def analyze_audio(self, audio_path):
        """Main analysis pipeline for ESC-50"""
        if not os.path.exists(audio_path):
            return {"error": "File not found"}

        print(f"\nAnalyzing: {os.path.basename(audio_path)}")

        # Initialize result dictionary
        result = {
            'sound_events': [],
            'transcription': "",
            'emotion': {'label': 'neutral', 'confidence': 0},
            'context': "",
            'priority': ""
        }

        # 1. Detect sound events
        result['sound_events'] = self.detect_sound_events(audio_path)

        # 2. Transcribe speech (though ESC-50 is mostly non-speech)
        try:
            transcription = self.stt_model.transcribe(audio_path)
            result['transcription'] = transcription["text"]
        except Exception as e:
            print(f"Error in transcription: {e}")
            result['transcription'] = ""

        # 3. Analyze emotion from transcription
        emotion_label, emotion_conf = self.analyze_emotion(result['transcription'])
        result['emotion'] = {
            'label': emotion_label,
            'confidence': emotion_conf
        }

        # 4. Generate insights
        insights = self.generate_insights(result['sound_events'], result['emotion'])
        result['context'] = insights['context']
        result['priority'] = insights['priority']

        return result

if __name__ == "__main__":
    train_new = input("Train new ESC-50 classification model? (y/n): ").strip().lower() == 'y'
    analyzer = ESC50Analyzer(train_new_model=train_new)

    # Get audio file path
    audio_path = input("Enter ESC-50 audio file path to analyze: ").strip()

    if os.path.exists(audio_path):
        result = analyzer.analyze_audio(audio_path)
        print(analyzer.format_output(result))

        # Optionally save results to JSON
        save_json = input("Save results to JSON file? (y/n): ").strip().lower()
        if save_json == 'y':
            with open('esc50_analysis_results.json', 'w') as f:
                json.dump(result, f, indent=2)
            print("Results saved to esc50_analysis_results.json")
    else:
        print("Error: File not found")
```

```python
import os
import random
import json
import librosa
import numpy as np
import pandas as pd
from collections import defaultdict
from transformers import pipeline
import whisper
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
from scipy import stats

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

class ESC50Analyzer:
    def __init__(self, train_new_model=False):
        # Load ESC-50 urgency mappings from CSV
        self.urgency_map = self.load_urgency_mappings('/content/drive/MyDrive/esc 50 meta data/ESC 50_10K_Urgenccy map.csv')

        # Initialize ESC-50 class mappings
        self.esc50_classes = [
            'dog', 'rooster', 'pig', 'cow', 'frog', 'cat', 'hen', 'insects', 'sheep', 'crow',
            'rain', 'sea_waves', 'crackling_fire', 'crickets', 'chirping_birds', 'water_drops',
            'wind', 'pouring_water', 'toilet_flush', 'thunderstorm', 'crying_baby', 'sneezing',
            'clapping', 'breathing', 'coughing', 'footsteps', 'laughing', 'brushing_teeth',
            'snoring', 'drinking_sipping', 'door_wood_knock', 'mouse_click', 'keyboard_typing',
            'door_wood_creaks', 'can_opening', 'washing_machine', 'vacuum_cleaner', 'clock_alarm',
            'clock_tick', 'glass_breaking', 'helicopter', 'chainsaw', 'siren', 'car_horn', 'engine',
            'train', 'church_bells', 'airplane', 'fireworks', 'hand_saw'
        ]

        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.esc50_classes)}
        self.idx_to_class = {idx: cls for idx, cls in enumerate(self.esc50_classes)}

        # Call initialize_models to setup models
        self.initialize_models(train_new_model)

    def load_urgency_mappings(self, csv_path):
        """Load urgency mappings from CSV with default fallback"""
        try:
            # Try to load from CSV (assuming format: Sound,Urgency)
            df = pd.read_csv(csv_path)
            return dict(zip(df['Sound'], df['Urgency']))
        except Exception as e:
            print(f"Error loading urgency mappings: {e}")
            # Default mappings for ESC-50 critical classes
            default_map = {
                'glass_breaking': 'high',
                'siren': 'high',
                'car_horn': 'high',
                'fireworks': 'high',
                'chainsaw': 'high',
                'helicopter': 'medium',
                'crying_baby': 'high',
                'thunderstorm': 'high',
                '_default': 'low'
            }
            print("Using default urgency mappings")
            return default_map

    def get_urgency(self, class_name):
        """Get urgency level for a class with fallback to default"""
        return self.urgency_map.get(class_name, self.urgency_map.get('_default', 'low'))

    def initialize_models(self, train_new_model):
        """Initialize sound classification and emotion analysis models"""
        # Initialize emotion classifier
        self.emotion_classifier = pipeline(
            "text-classification",
            model="j-hartmann/emotion-english-distilroberta-base",
            top_k=1
        )

        # Initialize speech-to-text model
        self.stt_model = whisper.load_model("base")

        self.scaler = StandardScaler()

        if train_new_model:
            print("Training new ESC-50 classification model...")
            self.train_sound_classifier()
        else:
            self.load_sound_model()

    def load_sound_model(self):
        """Try to load ESC-50 classification model"""
        model_paths = [
            "esc50_classifier.h5",
            "./esc50_classifier.h5",
            "/content/esc50_classifier.h5",
            "/content/drive/MyDrive/esc50_classifier.h5"
        ]

        for path in model_paths:
            try:
                self.sound_model = load_model(path)
                print(f"ESC-50 classification model loaded from {path}")
                # Load the corresponding scaler
                self.scaler = StandardScaler()
                self.scaler.mean_ = np.load("esc50_scaler_mean.npy")
                self.scaler.scale_ = np.load("esc50_scaler_scale.npy")
                print("Feature scaler loaded successfully")
                return
            except Exception as e:
                print(f"Error loading model from {path}: {str(e)}")
                continue

        print("ESC-50 classification model not found. You may want to train a new one.")

    def audio_augmentation(self, audio, sr):
        """Apply random audio augmentations"""
        # Time stretching
        rate = random.uniform(0.8, 1.2)
        audio = librosa.effects.time_stretch(audio, rate=rate)

        # Pitch shifting
        steps = random.randint(-2, 2)
        audio = librosa.effects.pitch_shift(audio, sr=sr, n_steps=steps)

        # Add noise
        noise = np.random.normal(0, 0.002, audio.shape)
        audio = audio + noise

        return audio * random.uniform(0.5, 1.5)

    def extract_features(self, audio, sample_rate):
        """Extract features optimized for ESC-50 dataset"""
        try:
            if sample_rate != 22050:
                audio = librosa.resample(audio, orig_sr=sample_rate, target_sr=22050)
                sample_rate = 22050

            features = []

            # MFCCs (50 coefficients for ESC-50)
            mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=50)
            mfccs_mean = np.mean(mfccs.T, axis=0)
            mfccs_var = np.var(mfccs.T, axis=0)
            features.extend(mfccs_mean)
            features.extend(mfccs_var)

            # Mel Spectrogram (128 bands)
            mel = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128)
            mel_mean = np.mean(mel.T, axis=0)
            mel_var = np.var(mel.T, axis=0)
            features.extend(mel_mean)
            features.extend(mel_var)

            # Chroma features
            chroma = librosa.feature.chroma_stft(y=audio, sr=sample_rate)
            chroma_mean = np.mean(chroma.T, axis=0)
            chroma_var = np.var(chroma.T, axis=0)
            features.extend(chroma_mean)
            features.extend(chroma_var)

            # Spectral features
            spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sample_rate)[0]
            spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sample_rate)[0]
            spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sample_rate)[0]
            spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sample_rate)

            features.extend([
                np.mean(spectral_centroid), np.var(spectral_centroid),
                np.mean(spectral_bandwidth), np.var(spectral_bandwidth),
                np.mean(spectral_rolloff), np.var(spectral_rolloff),
                np.mean(spectral_contrast), np.var(spectral_contrast)
            ])

            # Temporal features
            rms = librosa.feature.rms(y=audio)[0]
            zcr = librosa.feature.zero_crossing_rate(audio)[0]
            features.extend([np.mean(rms), np.var(rms), np.mean(zcr), np.var(zcr)])

            # Harmonic and percussive components
            harmonic = librosa.effects.harmonic(audio)
            percussive = librosa.effects.percussive(audio)
            features.extend([np.mean(harmonic), np.var(harmonic), np.mean(percussive), np.var(percussive)])

            return np.array(features)

        except Exception as e:
            print(f"Error extracting features: {e}")
            return None

    def load_esc50_data(self, dataset_path, max_samples_per_class=500):
        """Load ESC-50 dataset with augmentation"""
        features = []
        labels = []
        class_counts = defaultdict(int)

        print(f"Loading ESC-50 data from {dataset_path}")

        for class_name in self.esc50_classes:
            class_path = os.path.join(dataset_path, class_name)
            if not os.path.exists(class_path):
                print(f"Warning: Class folder '{class_name}' not found in dataset path")
                continue

            print(f"Processing {class_name}...")
            class_id = self.class_to_idx[class_name]

            audio_files = [f for f in os.listdir(class_path) if f.endswith(('.wav', '.mp3', '.ogg', '.flac'))]
            random.shuffle(audio_files)

            samples_loaded = 0
            for audio_file in audio_files:
                if samples_loaded >= max_samples_per_class:
                    break

                file_path = os.path.join(class_path, audio_file)

                try:
                    audio, sr = librosa.load(file_path, sr=None)

                    # ESC-50 files are 5 seconds long, pad/trim to match
                    target_length = 5 * sr
                    if len(audio) < target_length:
                        audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
                    else:
                        audio = audio[:target_length]

                    # Apply augmentation
                    augmented_audio = self.audio_augmentation(audio, sr)

                    # Extract features for both original and augmented audio
                    for a in [audio, augmented_audio]:
                        feature = self.extract_features(a, sr)
                        if feature is not None:
                            features.append(feature)
                            labels.append(class_id)
                            samples_loaded += 1

                            if samples_loaded >= max_samples_per_class:
                                break

                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")

        return np.array(features), np.array(labels)

    def plot_training_metrics(self, history):
        """Plot comprehensive training metrics with multiple visualization types"""
        plt.figure(figsize=(10, 6))

        # 1. Accuracy Curves (Linear and Semi-log)
        plt.subplot(2, 3, 1)
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy (Linear Scale)')
        plt.ylabel('Accuracy')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 3, 2)
        plt.semilogy(history.history['accuracy'], label='Train Accuracy')
        plt.semilogy(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy (Log Scale)')
        plt.ylabel('Accuracy (log)')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        # 2. Loss Curves (Linear and Semi-log)
        plt.subplot(2, 3, 3)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss (Linear Scale)')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 3, 4)
        plt.semilogy(history.history['loss'], label='Train Loss')
        plt.semilogy(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss (Log Scale)')
        plt.ylabel('Loss (log)')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        # 3. Learning Rate Decay (if available)
        if 'lr' in history.history:
            plt.subplot(2, 3, 5)
            plt.plot(history.history['lr'], label='Learning Rate')
            plt.title('Learning Rate Schedule')
            plt.ylabel('Learning Rate')
            plt.xlabel('Epoch')
            plt.legend()
            plt.grid(True)

        # 4. Error Rate (1 - accuracy)
        plt.subplot(2, 3, 6)
        train_error = [1 - acc for acc in history.history['accuracy']]
        val_error = [1 - acc for acc in history.history['val_accuracy']]
        plt.plot(train_error, label='Train Error Rate')
        plt.plot(val_error, label='Validation Error Rate')
        plt.title('Error Rate')
        plt.ylabel('Error Rate')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, y_true, y_pred):
        """Plot enhanced confusion matrix for ESC-50 classes"""
        cm = confusion_matrix(y_true, y_pred)
        cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

        plt.figure(figsize=(20, 20))

        # Plot raw counts
        plt.subplot(1, 2, 1)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.esc50_classes,
                    yticklabels=self.esc50_classes)
        plt.title('ESC-50 Confusion Matrix (Counts)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')

        # Plot normalized
        plt.subplot(1, 2, 2)
        sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                    xticklabels=self.esc50_classes,
                    yticklabels=self.esc50_classes)
        plt.title('ESC-50 Confusion Matrix (Normalized)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')

        plt.tight_layout()
        plt.show()

    def plot_class_distribution(self, y_true, y_pred=None):
        """Plot class distribution and prediction distribution"""
        plt.figure(figsize=(10, 6))

        # True class distribution
        plt.subplot(1, 3, 1)
        unique, counts = np.unique(y_true, return_counts=True)
        plt.bar([self.idx_to_class[u] for u in unique], counts)
        plt.title('True Class Distribution')
        plt.xticks(rotation=90)
        plt.ylabel('Count')

        if y_pred is not None:
            # Predicted class distribution
            plt.subplot(1, 3, 2)
            unique_pred, counts_pred = np.unique(y_pred, return_counts=True)
            plt.bar([self.idx_to_class[u] for u in unique_pred], counts_pred)
            plt.title('Predicted Class Distribution')
            plt.xticks(rotation=90)
            plt.ylabel('Count')

            # Error distribution
            plt.subplot(1, 3, 3)
            errors = y_pred != y_true
            error_classes = y_true[errors]
            unique_errors, counts_errors = np.unique(error_classes, return_counts=True)
            plt.bar([self.idx_to_class[u] for u in unique_errors], counts_errors)
            plt.title('Error Distribution by True Class')
            plt.xticks(rotation=90)
            plt.ylabel('Error Count')

        plt.tight_layout()
        plt.show()

    def plot_feature_importance(self, X_train):
        """Plot feature importance using correlation analysis"""
        # This is a simplified approach - for neural networks, consider using SHAP or permutation importance
        plt.figure(figsize=(10, 6))

        # Calculate correlation matrix (for continuous features)
        corr_matrix = np.corrcoef(X_train, rowvar=False)

        # Plot feature correlation heatmap
        sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
        plt.title('Feature Correlation Matrix')
        plt.xlabel('Feature Index')
        plt.ylabel('Feature Index')
        plt.show()

    def plot_confidence_distribution(self, y_true, y_pred_probs):
        """Plot confidence distribution for correct and incorrect predictions"""
        correct = y_pred_probs.max(axis=1) * (y_pred_probs.argmax(axis=1) == y_true)
        incorrect = y_pred_probs.max(axis=1) * (y_pred_probs.argmax(axis=1) != y_true)

        plt.figure(figsize=(12, 6))

        # Histogram of confidence scores
        plt.subplot(1, 2, 1)
        plt.hist(correct[correct > 0], bins=20, alpha=0.5, label='Correct', color='g')
        plt.hist(incorrect[incorrect > 0], bins=20, alpha=0.5, label='Incorrect', color='r')
        plt.title('Confidence Distribution')
        plt.xlabel('Confidence Score')
        plt.ylabel('Count')
        plt.legend()

        # ECDF plot
        plt.subplot(1, 2, 2)
        def ecdf(data):
            x = np.sort(data)
            y = np.arange(1, len(data)+1) / len(data)
            return x, y

        x_corr, y_corr = ecdf(correct[correct > 0])
        x_inc, y_inc = ecdf(incorrect[incorrect > 0])

        plt.plot(x_corr, y_corr, label='Correct', color='g')
        plt.plot(x_inc, y_inc, label='Incorrect', color='r')
        plt.title('ECDF of Confidence Scores')
        plt.xlabel('Confidence Score')
        plt.ylabel('ECDF')
        plt.legend()

        plt.tight_layout()
        plt.show()

    def train_sound_classifier(self):
        """Train new classifier on ESC-50 dataset with enhanced visualization"""
        dataset_path = input("Enter path to ESC-50 dataset folder (with class subfolders): ").strip()

        if not os.path.exists(dataset_path):
            print("Dataset folder not found.")
            return

        print("Loading ESC-50 dataset...")
        X, y = self.load_esc50_data(dataset_path)

        if len(X) == 0:
            print("No data loaded. Check dataset path and structure.")
            print("Expected folder structure: dataset_root/class_name/audio_files.wav")
            return

        # Plot initial class distribution
        self.plot_class_distribution(y)

        # Convert labels to categorical
        y_categorical = to_categorical(y, num_classes=len(self.esc50_classes))

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_categorical, test_size=0.2, random_state=42, stratify=y
        )

        # Scale features
        self.scaler = StandardScaler()
        X_train = self.scaler.fit_transform(X_train)
        X_test = self.scaler.transform(X_test)

        # Save scaler parameters
        np.save("esc50_scaler_mean.npy", self.scaler.mean_)
        np.save("esc50_scaler_scale.npy", self.scaler.scale_)

        # Plot feature importance
        self.plot_feature_importance(X_train)

        # Build model - deeper architecture for 50 classes
        model = Sequential([
            Dense(2048, activation='relu', input_shape=(X_train.shape[1],)),
            Dropout(0.5),
            Dense(1024, activation='relu'),
            Dropout(0.5),
            Dense(512, activation='relu'),
            Dropout(0.5),
            Dense(256, activation='relu'),
            Dropout(0.3),
            Dense(128, activation='relu'),
            Dropout(0.3),
            Dense(len(self.esc50_classes), activation='softmax')
        ])

        model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

        print("Training model...")
        history = model.fit(
            X_train, y_train,
            validation_data=(X_test, y_test),
            epochs=200,
            batch_size=128,
            callbacks=[early_stop],
            verbose=1
        )

        # Plot comprehensive training metrics
        self.plot_training_metrics(history)

        # Generate predictions for evaluation
        y_pred_probs = model.predict(X_test)
        y_pred_classes = np.argmax(y_pred_probs, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)

        # Plot confusion matrix
        self.plot_confusion_matrix(y_true_classes, y_pred_classes)

        # Plot class distribution and errors
        self.plot_class_distribution(y_true_classes, y_pred_classes)

        # Plot confidence distribution
        self.plot_confidence_distribution(y_true_classes, y_pred_probs)

        # Print classification report
        print("\nClassification Report:")
        print(classification_report(y_true_classes, y_pred_classes, target_names=self.esc50_classes))

        model.save("esc50_classifier.h5")
        self.sound_model = model
        print("ESC-50 model training completed.")

    def detect_sound_events(self, audio_path):
        """Detect sound events using ESC-50 model"""
        events = []
        try:
            audio, sr = librosa.load(audio_path, sr=None)

            # Ensure audio is 5 seconds long (ESC-50 standard)
            target_length = 5 * sr
            if len(audio) < target_length:
                audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
            else:
                audio = audio[:target_length]

            features = self.extract_features(audio, sr)
            if features is not None:
                features = self.scaler.transform(features.reshape(1, -1))
                prediction = self.sound_model.predict(features)
                predicted_class_idx = np.argmax(prediction, axis=1)[0]
                confidence = np.max(prediction)
                detected_class = self.idx_to_class.get(predicted_class_idx)

                if detected_class:
                    events.append({
                        'class': detected_class,
                        'confidence': float(confidence),
                        'urgency': self.get_urgency(detected_class)
                    })
        except Exception as e:
            print(f"Error in sound event detection: {e}")
        return events

    def analyze_emotion(self, text):
        """Analyze emotion from text"""
        if not text.strip():
            return "neutral", 0.0

        try:
            result = self.emotion_classifier(text)
            if result and len(result) > 0:
                emotion = result[0][0]['label']
                confidence = result[0][0]['score']
                return emotion, float(confidence)
        except Exception as e:
            print(f"Error in emotion analysis: {e}")

        return "neutral", 0.0

    def format_output(self, analysis_result):
        """Format the analysis results for display"""
        output = "\n=== ESC-50 Audio Analysis Results ===\n"

        # Sound events
        output += "\n🔊 Detected Sound Events:\n"
        if analysis_result['sound_events']:
            for event in analysis_result['sound_events']:
                output += (f"- {event['class'].replace('_', ' ').title()} "
                          f"[Confidence: {event['confidence']:.2f}, "
                          f"Urgency: {event['urgency'].capitalize()}]\n")
        else:
            output += "No sound events detected\n"

        # Transcription
        if analysis_result['transcription']:
            output += f"\n🗣️ Transcription: \"{analysis_result['transcription']}\""
            output += f"\n😃 Emotion: {analysis_result['emotion']['label'].capitalize()} "
            output += f"(Confidence: {analysis_result['emotion']['confidence']:.2f})\n"
        else:
            output += "\nNo speech detected\n"

        # Insights
        output += f"\n💡 Actionable Insights:\n"
        output += f"- Context: {analysis_result['context']}\n"
        output += f"- Priority: {analysis_result['priority']}\n"

        return output

    def generate_insights(self, sound_events, emotion_result):
        """Generate actionable insights from analysis"""
        insights = []
        context_parts = []
        priority = "Low"

        # Sound-based insights
        if sound_events:
            high_urgency = [e for e in sound_events if e['urgency'] == 'high']
            medium_urgency = [e for e in sound_events if e['urgency'] == 'medium']

            if high_urgency:
                sounds = ", ".join(set(e['class'].replace('_', ' ').title() for e in high_urgency))
                context_parts.append(f"High urgency sounds detected: {sounds}")
                priority = "High"

            if medium_urgency:
                sounds = ", ".join(set(e['class'].replace('_', ' ').title() for e in medium_urgency))
                context_parts.append(f"Medium urgency sounds detected: {sounds}")
                if priority != "High":
                    priority = "Medium"

        # Emotion-based insights
        emotion = emotion_result['label']
        confidence = emotion_result['confidence']

        if confidence > 0.7:
            if emotion in ['anger', 'fear']:
                context_parts.append(f"Strong negative emotion detected ({emotion})")
                if priority != "High":
                    priority = "Medium"
            elif emotion == 'joy':
                context_parts.append("Positive emotional state detected")

        if not context_parts:
            context_parts.append("No significant events detected")

        context = ". ".join(context_parts) + "."

        # Enhanced priority determination
        if priority == "High":
            priority = "Immediate attention required"
        elif priority == "Medium":
            priority = "Monitoring recommended"
        else:
            priority = "No immediate action needed"

        return {
            'context': context,
            'priority': priority
        }

    def analyze_audio(self, audio_path):
        """Main analysis pipeline for ESC-50"""
        if not os.path.exists(audio_path):
            return {"error": "File not found"}

        print(f"\nAnalyzing: {os.path.basename(audio_path)}")

        # Initialize result dictionary
        result = {
            'sound_events': [],
            'transcription': "",
            'emotion': {'label': 'neutral', 'confidence': 0},
            'context': "",
            'priority': ""
        }

        # 1. Detect sound events
        result['sound_events'] = self.detect_sound_events(audio_path)

        # 2. Transcribe speech (though ESC-50 is mostly non-speech)
        try:
            transcription = self.stt_model.transcribe(audio_path)
            result['transcription'] = transcription["text"]
        except Exception as e:
            print(f"Error in transcription: {e}")
            result['transcription'] = ""

        # 3. Analyze emotion from transcription
        emotion_label, emotion_conf = self.analyze_emotion(result['transcription'])
        result['emotion'] = {
            'label': emotion_label,
            'confidence': emotion_conf
        }

        # 4. Generate insights
        insights = self.generate_insights(result['sound_events'], result['emotion'])
        result['context'] = insights['context']
        result['priority'] = insights['priority']

        return result

    def plot_comparative_results(self, results):
        """Plot comparative analysis of multiple audio files"""
        # Prepare data for plotting
        files = []
        sound_counts = []
        urgency_levels = {'high': [], 'medium': [], 'low': []}
        emotions = []
        emotion_confidences = []

        for i, result in enumerate(results):
            files.append(f"Audio {i+1}")
            sound_counts.append(len(result['sound_events']))

            # Count urgency levels
            urgency_counts = {'high': 0, 'medium': 0, 'low': 0}
            for event in result['sound_events']:
                urgency_counts[event['urgency']] += 1

            for level in urgency_levels:
                urgency_levels[level].append(urgency_counts[level])

            # Emotion data
            emotions.append(result['emotion']['label'])
            emotion_confidences.append(result['emotion']['confidence'])

        # Create figure
        plt.figure(figsize=(10, 6))

        # 1. Sound Event Count
        plt.subplot(2, 2, 1)
        plt.bar(files, sound_counts, color='skyblue')
        plt.title('Number of Sound Events Detected')
        plt.ylabel('Count')

        # 2. Urgency Level Distribution
        plt.subplot(2, 2, 2)
        bottom = np.zeros(len(files))
        for level, counts in urgency_levels.items():
            plt.bar(files, counts, bottom=bottom, label=level.capitalize())
            bottom += counts
        plt.title('Urgency Level Distribution')
        plt.ylabel('Count')
        plt.legend()

        # 3. Emotion Distribution
        plt.subplot(2, 2, 3)
        unique_emotions = list(set(emotions))
        emotion_counts = [emotions.count(e) for e in unique_emotions]
        plt.bar(unique_emotions, emotion_counts, color='lightgreen')
        plt.title('Emotion Distribution Across Files')
        plt.ylabel('Count')

        # 4. Emotion Confidence
        plt.subplot(2, 2, 4)
        plt.bar(files, emotion_confidences, color='salmon')
        plt.title('Emotion Detection Confidence')
        plt.ylabel('Confidence Score')
        plt.ylim(0, 1)

        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    train_new = input("Train new ESC-50 classification model? (y/n): ").strip().lower() == 'y'
    analyzer = ESC50Analyzer(train_new_model=train_new)

    # Get multiple audio file paths
    audio_paths = []
    for i in range(5):
        path = input(f"Enter ESC-50 audio file path {i+1}/5 to analyze: ").strip()
        if os.path.exists(path):
            audio_paths.append(path)
        else:
            print(f"Warning: File not found - {path}")

    if not audio_paths:
        print("Error: No valid audio files provided")
    else:
        results = []
        for path in audio_paths:
            print(f"\nAnalyzing: {os.path.basename(path)}")
            result = analyzer.analyze_audio(path)
            print(analyzer.format_output(result))
            results.append(result)

        # Plot comparative results
        analyzer.plot_comparative_results(results)

        # Optionally save results to JSON
        save_json = input("Save results to JSON file? (y/n): ").strip().lower()
        if save_json == 'y':
            with open('esc50_analysis_results.json', 'w') as f:
                json.dump(results, f, indent=2)
            print("Results saved to esc50_analysis_results.json")
```

```python
import os
import random
import json
import librosa
import numpy as np
import pandas as pd
from collections import defaultdict
from transformers import pipeline
import whisper
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
from scipy import stats

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

class ESC50Analyzer:
    def __init__(self, train_new_model=False):
        # Load ESC-50 urgency mappings from CSV
        self.urgency_map = self.load_urgency_mappings('/content/ESC 50_10K_Urgenccy map.csv')

        # Initialize ESC-50 class mappings
        self.esc50_classes = [
            'dog', 'rooster', 'pig', 'cow', 'frog', 'cat', 'hen', 'insects', 'sheep', 'crow',
            'rain', 'sea_waves', 'crackling_fire', 'crickets', 'chirping_birds', 'water_drops',
            'wind', 'pouring_water', 'toilet_flush', 'thunderstorm', 'crying_baby', 'sneezing',
            'clapping', 'breathing', 'coughing', 'footsteps', 'laughing', 'brushing_teeth',
            'snoring', 'drinking_sipping', 'door_wood_knock', 'mouse_click', 'keyboard_typing',
            'door_wood_creaks', 'can_opening', 'washing_machine', 'vacuum_cleaner', 'clock_alarm',
            'clock_tick', 'glass_breaking', 'helicopter', 'chainsaw', 'siren', 'car_horn', 'engine',
            'train', 'church_bells', 'airplane', 'fireworks', 'hand_saw'
        ]

        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.esc50_classes)}
        self.idx_to_class = {idx: cls for idx, cls in enumerate(self.esc50_classes)}

        # Call initialize_models to setup models
        self.initialize_models(train_new_model)

    def load_urgency_mappings(self, csv_path):
        """Load urgency mappings from CSV with default fallback"""
        try:
            # Try to load from CSV (assuming format: Sound,Urgency)
            df = pd.read_csv(csv_path)
            return dict(zip(df['class'], df['urgency']))
        except Exception as e:
            print(f"Error loading urgency mappings: {e}")
            # Default mappings for ESC-50 critical classes
            default_map = {
                'glass_breaking': 'high',
                'siren': 'high',
                'car_horn': 'high',
                'fireworks': 'high',
                'chainsaw': 'high',
                'helicopter': 'medium',
                'crying_baby': 'high',
                'thunderstorm': 'high',
                '_default': 'low'
            }
            print("Using default urgency mappings")
            return default_map

    def get_urgency(self, class_name):
        """Get urgency level for a class with fallback to default"""
        return self.urgency_map.get(class_name, self.urgency_map.get('_default', 'low'))

    def initialize_models(self, train_new_model):
        """Initialize sound classification and emotion analysis models"""
        # Initialize emotion classifier
        self.emotion_classifier = pipeline(
            "text-classification",
            model="j-hartmann/emotion-english-distilroberta-base",
            top_k=1
        )

        # Initialize speech-to-text model
        self.stt_model = whisper.load_model("base")

        self.scaler = StandardScaler()

        if train_new_model:
            print("Training new ESC-50 classification model...")
            self.train_sound_classifier()
        else:
            self.load_sound_model()

    def load_sound_model(self):
        """Try to load ESC-50 classification model"""
        model_paths = [
            "esc50_classifier.h5",
            "./esc50_classifier.h5",
            "/content/esc50_classifier.h5",
            "/content/drive/MyDrive/esc50_classifier.h5"
        ]

        for path in model_paths:
            try:
                self.sound_model = load_model(path)
                print(f"ESC-50 classification model loaded from {path}")
                # Load the corresponding scaler
                self.scaler = StandardScaler()
                self.scaler.mean_ = np.load("esc50_scaler_mean.npy")
                self.scaler.scale_ = np.load("esc50_scaler_scale.npy")
                print("Feature scaler loaded successfully")
                return
            except Exception as e:
                print(f"Error loading model from {path}: {str(e)}")
                continue

        print("ESC-50 classification model not found. You may want to train a new one.")

    def audio_augmentation(self, audio, sr):
        """Apply random audio augmentations"""
        # Time stretching
        rate = random.uniform(0.8, 1.2)
        audio = librosa.effects.time_stretch(audio, rate=rate)

        # Pitch shifting
        steps = random.randint(-2, 2)
        audio = librosa.effects.pitch_shift(audio, sr=sr, n_steps=steps)

        # Add noise
        noise = np.random.normal(0, 0.002, audio.shape)
        audio = audio + noise

        return audio * random.uniform(0.5, 1.5)

    def extract_features(self, audio, sample_rate):
        """Extract features optimized for ESC-50 dataset"""
        try:
            if sample_rate != 22050:
                audio = librosa.resample(audio, orig_sr=sample_rate, target_sr=22050)
                sample_rate = 22050

            features = []

            # MFCCs (50 coefficients for ESC-50)
            mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=50)
            mfccs_mean = np.mean(mfccs.T, axis=0)
            mfccs_var = np.var(mfccs.T, axis=0)
            features.extend(mfccs_mean)
            features.extend(mfccs_var)

            # Mel Spectrogram (128 bands)
            mel = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=128)
            mel_mean = np.mean(mel.T, axis=0)
            mel_var = np.var(mel.T, axis=0)
            features.extend(mel_mean)
            features.extend(mel_var)

            # Chroma features
            chroma = librosa.feature.chroma_stft(y=audio, sr=sample_rate)
            chroma_mean = np.mean(chroma.T, axis=0)
            chroma_var = np.var(chroma.T, axis=0)
            features.extend(chroma_mean)
            features.extend(chroma_var)

            # Spectral features
            spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sample_rate)[0]
            spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sample_rate)[0]
            spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sample_rate)[0]
            spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sample_rate)

            features.extend([
                np.mean(spectral_centroid), np.var(spectral_centroid),
                np.mean(spectral_bandwidth), np.var(spectral_bandwidth),
                np.mean(spectral_rolloff), np.var(spectral_rolloff),
                np.mean(spectral_contrast), np.var(spectral_contrast)
            ])

            # Temporal features
            rms = librosa.feature.rms(y=audio)[0]
            zcr = librosa.feature.zero_crossing_rate(audio)[0]
            features.extend([np.mean(rms), np.var(rms), np.mean(zcr), np.var(zcr)])

            # Harmonic and percussive components
            harmonic = librosa.effects.harmonic(audio)
            percussive = librosa.effects.percussive(audio)
            features.extend([np.mean(harmonic), np.var(harmonic), np.mean(percussive), np.var(percussive)])

            return np.array(features)

        except Exception as e:
            print(f"Error extracting features: {e}")
            return None

    def load_esc50_data(self, dataset_path, max_samples_per_class=500):
        """Load ESC-50 dataset with augmentation"""
        features = []
        labels = []
        class_counts = defaultdict(int)

        print(f"Loading ESC-50 data from {dataset_path}")

        for class_name in self.esc50_classes:
            class_path = os.path.join(dataset_path, class_name)
            if not os.path.exists(class_path):
                print(f"Warning: Class folder '{class_name}' not found in dataset path")
                continue

            print(f"Processing {class_name}...")
            class_id = self.class_to_idx[class_name]

            audio_files = [f for f in os.listdir(class_path) if f.endswith(('.wav', '.mp3', '.ogg', '.flac'))]
            random.shuffle(audio_files)

            samples_loaded = 0
            for audio_file in audio_files:
                if samples_loaded >= max_samples_per_class:
                    break

                file_path = os.path.join(class_path, audio_file)

                try:
                    audio, sr = librosa.load(file_path, sr=None)

                    # ESC-50 files are 5 seconds long, pad/trim to match
                    target_length = 5 * sr
                    if len(audio) < target_length:
                        audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
                    else:
                        audio = audio[:target_length]

                    # Apply augmentation
                    augmented_audio = self.audio_augmentation(audio, sr)

                    # Extract features for both original and augmented audio
                    for a in [audio, augmented_audio]:
                        feature = self.extract_features(a, sr)
                        if feature is not None:
                            features.append(feature)
                            labels.append(class_id)
                            samples_loaded += 1

                            if samples_loaded >= max_samples_per_class:
                                break

                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")

        return np.array(features), np.array(labels)

    def plot_training_metrics(self, history):
        """Plot comprehensive training metrics with multiple visualization types"""
        plt.figure(figsize=(10, 6))

        # 1. Accuracy Curves (Linear and Semi-log)
        plt.subplot(2, 3, 1)
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy (Linear Scale)')
        plt.ylabel('Accuracy')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 3, 2)
        plt.semilogy(history.history['accuracy'], label='Train Accuracy')
        plt.semilogy(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy (Log Scale)')
        plt.ylabel('Accuracy (log)')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        # 2. Loss Curves (Linear and Semi-log)
        plt.subplot(2, 3, 3)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss (Linear Scale)')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 3, 4)
        plt.semilogy(history.history['loss'], label='Train Loss')
        plt.semilogy(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss (Log Scale)')
        plt.ylabel('Loss (log)')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        # 3. Learning Rate Decay (if available)
        if 'lr' in history.history:
            plt.subplot(2, 3, 5)
            plt.plot(history.history['lr'], label='Learning Rate')
            plt.title('Learning Rate Schedule')
            plt.ylabel('Learning Rate')
            plt.xlabel('Epoch')
            plt.legend()
            plt.grid(True)

        # 4. Error Rate (1 - accuracy)
        plt.subplot(2, 3, 6)
        train_error = [1 - acc for acc in history.history['accuracy']]
        val_error = [1 - acc for acc in history.history['val_accuracy']]
        plt.plot(train_error, label='Train Error Rate')
        plt.plot(val_error, label='Validation Error Rate')
        plt.title('Error Rate')
        plt.ylabel('Error Rate')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, y_true, y_pred):
        """Plot enhanced confusion matrix for ESC-50 classes"""
        cm = confusion_matrix(y_true, y_pred)
        cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

        plt.figure(figsize=(20, 20))

        # Plot raw counts
        plt.subplot(1, 2, 1)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.esc50_classes,
                    yticklabels=self.esc50_classes)
        plt.title('ESC-50 Confusion Matrix (Counts)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')

        # Plot normalized
        plt.subplot(1, 2, 2)
        sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                    xticklabels=self.esc50_classes,
                    yticklabels=self.esc50_classes)
        plt.title('ESC-50 Confusion Matrix (Normalized)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')

        plt.tight_layout()
        plt.show()

    def plot_class_distribution(self, y_true, y_pred=None):
        """Plot class distribution and prediction distribution"""
        plt.figure(figsize=(10, 6))

        # True class distribution
        plt.subplot(1, 3, 1)
        unique, counts = np.unique(y_true, return_counts=True)
        plt.bar([self.idx_to_class[u] for u in unique], counts)
        plt.title('True Class Distribution')
        plt.xticks(rotation=90)
        plt.ylabel('Count')

        if y_pred is not None:
            # Predicted class distribution
            plt.subplot(1, 3, 2)
            unique_pred, counts_pred = np.unique(y_pred, return_counts=True)
            plt.bar([self.idx_to_class[u] for u in unique_pred], counts_pred)
            plt.title('Predicted Class Distribution')
            plt.xticks(rotation=90)
            plt.ylabel('Count')

            # Error distribution
            plt.subplot(1, 3, 3)
            errors = y_pred != y_true
            error_classes = y_true[errors]
            unique_errors, counts_errors = np.unique(error_classes, return_counts=True)
            plt.bar([self.idx_to_class[u] for u in unique_errors], counts_errors)
            plt.title('Error Distribution by True Class')
            plt.xticks(rotation=90)
            plt.ylabel('Error Count')

        plt.tight_layout()
        plt.show()

    def plot_feature_importance(self, X_train):
        """Plot feature importance using correlation analysis"""
        # This is a simplified approach - for neural networks, consider using SHAP or permutation importance
        plt.figure(figsize=(10, 6))

        # Calculate correlation matrix (for continuous features)
        corr_matrix = np.corrcoef(X_train, rowvar=False)

        # Plot feature correlation heatmap
        sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
        plt.title('Feature Correlation Matrix')
        plt.xlabel('Feature Index')
        plt.ylabel('Feature Index')
        plt.show()

    def plot_confidence_distribution(self, y_true, y_pred_probs):
        """Plot confidence distribution for correct and incorrect predictions"""
        correct = y_pred_probs.max(axis=1) * (y_pred_probs.argmax(axis=1) == y_true)
        incorrect = y_pred_probs.max(axis=1) * (y_pred_probs.argmax(axis=1) != y_true)

        plt.figure(figsize=(10, 6))

        # Histogram of confidence scores
        plt.subplot(1, 2, 1)
        plt.hist(correct[correct > 0], bins=20, alpha=0.5, label='Correct', color='g')
        plt.hist(incorrect[incorrect > 0], bins=20, alpha=0.5, label='Incorrect', color='r')
        plt.title('Confidence Distribution')
        plt.xlabel('Confidence Score')
        plt.ylabel('Count')
        plt.legend()

        # ECDF plot
        plt.subplot(1, 2, 2)
        def ecdf(data):
            x = np.sort(data)
            y = np.arange(1, len(data)+1) / len(data)
            return x, y

        x_corr, y_corr = ecdf(correct[correct > 0])
        x_inc, y_inc = ecdf(incorrect[incorrect > 0])

        plt.plot(x_corr, y_corr, label='Correct', color='g')
        plt.plot(x_inc, y_inc, label='Incorrect', color='r')
        plt.title('ECDF of Confidence Scores')
        plt.xlabel('Confidence Score')
        plt.ylabel('ECDF')
        plt.legend()

        plt.tight_layout()
        plt.show()

    def train_sound_classifier(self):
        """Train new classifier on ESC-50 dataset with enhanced visualization"""
        dataset_path = input("Enter path to ESC-50 dataset folder (with class subfolders): ").strip()

        if not os.path.exists(dataset_path):
            print("Dataset folder not found.")
            return

        print("Loading ESC-50 dataset...")
        X, y = self.load_esc50_data(dataset_path)

        if len(X) == 0:
            print("No data loaded. Check dataset path and structure.")
            print("Expected folder structure: dataset_root/class_name/audio_files.wav")
            return

        # Plot initial class distribution
        self.plot_class_distribution(y)

        # Convert labels to categorical
        y_categorical = to_categorical(y, num_classes=len(self.esc50_classes))

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_categorical, test_size=0.2, random_state=42, stratify=y
        )

        # Scale features
        self.scaler = StandardScaler()
        X_train = self.scaler.fit_transform(X_train)
        X_test = self.scaler.transform(X_test)

        # Save scaler parameters
        np.save("esc50_scaler_mean.npy", self.scaler.mean_)
        np.save("esc50_scaler_scale.npy", self.scaler.scale_)

        # Plot feature importance
        self.plot_feature_importance(X_train)

        # Build model - deeper architecture for 50 classes
        model = Sequential([
            Dense(2048, activation='relu', input_shape=(X_train.shape[1],)),
            Dropout(0.5),
            Dense(1024, activation='relu'),
            Dropout(0.5),
            Dense(512, activation='relu'),
            Dropout(0.5),
            Dense(256, activation='relu'),
            Dropout(0.3),
            Dense(128, activation='relu'),
            Dropout(0.3),
            Dense(len(self.esc50_classes), activation='softmax')
        ])

        model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

        print("Training model...")
        history = model.fit(
            X_train, y_train,
            validation_data=(X_test, y_test),
            epochs=200,
            batch_size=128,
            callbacks=[early_stop],
            verbose=1
        )

        # Plot comprehensive training metrics
        self.plot_training_metrics(history)

        # Generate predictions for evaluation
        y_pred_probs = model.predict(X_test)
        y_pred_classes = np.argmax(y_pred_probs, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)

        # Plot confusion matrix
        self.plot_confusion_matrix(y_true_classes, y_pred_classes)

        # Plot class distribution and errors
        self.plot_class_distribution(y_true_classes, y_pred_classes)

        # Plot confidence distribution
        self.plot_confidence_distribution(y_true_classes, y_pred_probs)

        # Print classification report
        print("\nClassification Report:")
        print(classification_report(y_true_classes, y_pred_classes, target_names=self.esc50_classes))

        model.save("esc50_classifier.h5")
        self.sound_model = model
        print("ESC-50 model training completed.")

    def detect_sound_events(self, audio_path):
        """Detect sound events using ESC-50 model"""
        events = []
        try:
            audio, sr = librosa.load(audio_path, sr=None)

            # Ensure audio is 5 seconds long (ESC-50 standard)
            target_length = 5 * sr
            if len(audio) < target_length:
                audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
            else:
                audio = audio[:target_length]

            features = self.extract_features(audio, sr)
            if features is not None:
                features = self.scaler.transform(features.reshape(1, -1))
                prediction = self.sound_model.predict(features)
                predicted_class_idx = np.argmax(prediction, axis=1)[0]
                confidence = np.max(prediction)
                detected_class = self.idx_to_class.get(predicted_class_idx)

                if detected_class:
                    events.append({
                        'class': detected_class,
                        'confidence': float(confidence),
                        'urgency': self.get_urgency(detected_class)
                    })
        except Exception as e:
            print(f"Error in sound event detection: {e}")
        return events

    def analyze_emotion(self, text):
        """Analyze emotion from text"""
        if not text.strip():
            return "neutral", 0.0

        try:
            result = self.emotion_classifier(text)
            if result and len(result) > 0:
                emotion = result[0][0]['label']
                confidence = result[0][0]['score']
                return emotion, float(confidence)
        except Exception as e:
            print(f"Error in emotion analysis: {e}")

        return "neutral", 0.0

    def format_output(self, analysis_result):
        """Format the analysis results for display"""
        output = "\n=== ESC-50 Audio Analysis Results ===\n"

        # Sound events
        output += "\n🔊 Detected Sound Events:\n"
        if analysis_result['sound_events']:
            for event in analysis_result['sound_events']:
                output += (f"- {event['class'].replace('_', ' ').title()} "
                          f"[Confidence: {event['confidence']:.2f}, "
                          f"Urgency: {event['urgency'].capitalize()}]\n")
        else:
            output += "No sound events detected\n"

        # Transcription
        if analysis_result['transcription']:
            output += f"\n🗣️ Transcription: \"{analysis_result['transcription']}\""
            output += f"\n😃 Emotion: {analysis_result['emotion']['label'].capitalize()} "
            output += f"(Confidence: {analysis_result['emotion']['confidence']:.2f})\n"
        else:
            output += "\nNo speech detected\n"

        # Insights
        output += f"\n💡 Actionable Insights:\n"
        output += f"- Context: {analysis_result['context']}\n"
        output += f"- Priority: {analysis_result['priority']}\n"

        return output

    def generate_insights(self, sound_events, emotion_result):
        """Generate actionable insights from analysis"""
        insights = []
        context_parts = []
        priority = "Low"

        # Sound-based insights
        if sound_events:
            high_urgency = [e for e in sound_events if e['urgency'] == 'high']
            medium_urgency = [e for e in sound_events if e['urgency'] == 'medium']

            if high_urgency:
                sounds = ", ".join(set(e['class'].replace('_', ' ').title() for e in high_urgency))
                context_parts.append(f"High urgency sounds detected: {sounds}")
                priority = "High"

            if medium_urgency:
                sounds = ", ".join(set(e['class'].replace('_', ' ').title() for e in medium_urgency))
                context_parts.append(f"Medium urgency sounds detected: {sounds}")
                if priority != "High":
                    priority = "Medium"

        # Emotion-based insights
        emotion = emotion_result['label']
        confidence = emotion_result['confidence']

        if confidence > 0.7:
            if emotion in ['anger', 'fear']:
                context_parts.append(f"Strong negative emotion detected ({emotion})")
                if priority != "High":
                    priority = "Medium"
            elif emotion == 'joy':
                context_parts.append("Positive emotional state detected")

        if not context_parts:
            context_parts.append("No significant events detected")

        context = ". ".join(context_parts) + "."

        # Enhanced priority determination
        if priority == "High":
            priority = "Immediate attention required"
        elif priority == "Medium":
            priority = "Monitoring recommended"
        else:
            priority = "No immediate action needed"

        return {
            'context': context,
            'priority': priority
        }

    def analyze_audio(self, audio_path):
        """Main analysis pipeline for ESC-50"""
        if not os.path.exists(audio_path):
            return {"error": "File not found"}

        print(f"\nAnalyzing: {os.path.basename(audio_path)}")

        # Initialize result dictionary
        result = {
            'sound_events': [],
            'transcription': "",
            'emotion': {'label': 'neutral', 'confidence': 0},
            'context': "",
            'priority': ""
        }

        # 1. Detect sound events
        result['sound_events'] = self.detect_sound_events(audio_path)

        # 2. Transcribe speech (though ESC-50 is mostly non-speech)
        try:
            transcription = self.stt_model.transcribe(audio_path)
            result['transcription'] = transcription["text"]
        except Exception as e:
            print(f"Error in transcription: {e}")
            result['transcription'] = ""

        # 3. Analyze emotion from transcription
        emotion_label, emotion_conf = self.analyze_emotion(result['transcription'])
        result['emotion'] = {
            'label': emotion_label,
            'confidence': emotion_conf
        }

        # 4. Generate insights
        insights = self.generate_insights(result['sound_events'], result['emotion'])
        result['context'] = insights['context']
        result['priority'] = insights['priority']

        return result

    def plot_comparative_results(self, results):
        """Plot comparative analysis of multiple audio files"""
        # Prepare data for plotting
        files = []
        sound_counts = []
        urgency_levels = {'high': [], 'medium': [], 'low': []}
        emotions = []
        emotion_confidences = []

        for i, result in enumerate(results):
            files.append(f"Audio {i+1}")
            sound_counts.append(len(result['sound_events']))

            # Count urgency levels
            urgency_counts = {'high': 0, 'medium': 0, 'low': 0}
            for event in result['sound_events']:
                urgency_counts[event['urgency']] += 1

            for level in urgency_levels:
                urgency_levels[level].append(urgency_counts[level])

            # Emotion data
            emotions.append(result['emotion']['label'])
            emotion_confidences.append(result['emotion']['confidence'])

        # Create figure
        plt.figure(figsize=(10, 6))

        # 1. Sound Event Count
        plt.subplot(2, 2, 1)
        plt.bar(files, sound_counts, color='skyblue')
        plt.title('Number of Sound Events Detected')
        plt.ylabel('Count')

        # 2. Urgency Level Distribution
        plt.subplot(2, 2, 2)
        bottom = np.zeros(len(files))
        for level, counts in urgency_levels.items():
            plt.bar(files, counts, bottom=bottom, label=level.capitalize())
            bottom += counts
        plt.title('Urgency Level Distribution')
        plt.ylabel('Count')
        plt.legend()

        # 3. Emotion Distribution
        plt.subplot(2, 2, 3)
        unique_emotions = list(set(emotions))
        emotion_counts = [emotions.count(e) for e in unique_emotions]
        plt.bar(unique_emotions, emotion_counts, color='lightgreen')
        plt.title('Emotion Distribution Across Files')
        plt.ylabel('Count')

        # 4. Emotion Confidence
        plt.subplot(2, 2, 4)
        plt.bar(files, emotion_confidences, color='salmon')
        plt.title('Emotion Detection Confidence')
        plt.ylabel('Confidence Score')
        plt.ylim(0, 1)

        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    train_new = input("Train new ESC-50 classification model? (y/n): ").strip().lower() == 'y'
    analyzer = ESC50Analyzer(train_new_model=train_new)

    # Get multiple audio file paths
    audio_paths = []
    for i in range(5):
        path = input(f"Enter ESC-50 audio file path {i+1}/5 to analyze: ").strip()
        if os.path.exists(path):
            audio_paths.append(path)
        else:
            print(f"Warning: File not found - {path}")

    if not audio_paths:
        print("Error: No valid audio files provided")
    else:
        results = []
        for path in audio_paths:
            print(f"\nAnalyzing: {os.path.basename(path)}")
            result = analyzer.analyze_audio(path)
            print(analyzer.format_output(result))
            results.append(result)

        # Plot comparative results
        analyzer.plot_comparative_results(results)

        # Optionally save results to JSON
        save_json = input("Save results to JSON file? (y/n): ").strip().lower()
        if save_json == 'y':
            with open('esc50_analysis_results.json', 'w') as f:
                json.dump(results, f, indent=2)
            print("Results saved to esc50_analysis_results.json")
```
